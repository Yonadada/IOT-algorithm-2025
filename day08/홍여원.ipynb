```python
## 전역 변수 ##
SIZE = 5
queue = [None for _ in range(SIZE)] 
front = rear = -1 # 뒤에서 1개씩 뺀다

## 함수 선언 부분 ##
# 큐가 가득 찼을 때
def isQueueFull():
    
    ## 전역변수 사용
    global SIZE, queue, front, rear
    if rear == SIZE-1 :
        return True 
    else:
        return False

## 큐가 다 비었을 때     
def isQueueEmpty():
    global SIZE, queue, front, rear
    # 제일 앞과 뒤가 같을 때 즉, 없을 때 
    if front == rear :
        return True
    else: 
        # 비어있지 않을 때 
        return False

## 큐에 데이터 추가시
def enQueue(data):    
    global SIZE, queue, front, rear
    # 만약 큐가 다 찼을 경우
    if isQueueFull():
        print('자리 다 차서 들어갈 자리 없음')
        return
    # 자리가 있을 시 맨 뒤에 데이터를 적재
    rear += 1 
    queue[rear] = data
    

## 큐에 있는 데이터를 삭제할 시
def deQueue(data):
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 다 비어져있습니다')
        return None
    front += 1
    data = queue[front]
    queue[front] = None
    
    # 삭제한 부분 공백 없애게 칸 앞으로 땡기기
    for i in range(front +1 , rear + 1):
        queue[i-1] = queue[i]
        queue[i] = None
    front =-1
    rear -= 1
    
    return data
    
##  데이터 꺼내기 
def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 비어져 있습니다')
        return None

    return queue[front + 1]
        
        
        
## 메인 부분 ## 
if __name__ == "__main__":
    enQueue('정국')
    enQueue('뷔')
    enQueue('지민')
    enQueue('진')
    enQueue('슈가')
    print(f'대기줄 상태 : {queue}')
    

    

    # 데이터 추가시 맨뒤로 적재
    for _ in range(rear +1 ) :
        
        print(f'{deQueue} 님 식당에 들어감')
        print(f'대기줄 상태 : {queue}')
    
    print('식당 영업 종료!')
```

3. 편의점 판매된 물건

```python
import random

## 전역 변수 ##
memory = []
root = None # 첫번째 데이터는 None
objectAry = ['바나나맛 우유', '레쓰비캔커피','츄파츕스','도시락','삼다수','코카콜라','삼각김밥']
sellAry = [random.choice(objectAry) for _ in range(20)] # ohjectAry에서 20개를 랜덤으로 골라 sellAry[] 

print(f'오늘 판매된 물건(중복O) --> {sellAry}')

## 함수 선언 ##
def preorder(node):
    if node == None:
        return 
    
    # node에 있는 데이터 출력 
    print(node.data, end=' ')
    # 왼쪽 노드 
    preorder(node.left)
    # 오른쪽 노드
    preorder(node.right)



## 메인 ## 
class TreeNode(): # TreeNode 클래스 생성
    def __init__(self): # 자기자신 초기화 
        self.left = None # 자신의 노드보다 작으면 왼쪽 크면 오른쪽 
        self.right = None
        self.data = None
        
        
node = TreeNode()
node.data = sellAry[0] # 데이터 삽입시 0번째로
root = node 
memory.append(node) # memomory 리스트에 node 적재

for name in sellAry[1:]: # 처음부터 끝까지 반복  
    
    node  = TreeNode()
    node.data = name 
    
    current = root 
    while True :
        if name == current.data: # 추가할 데이터와 현재 데이터가 중복되면 -> 아!! 판매된 물품일경우 brek
            break
        
        if name < current.data: # 그 노드에 있는 숫자보다 name이 작다면 
            if current.left == None: # 노드 왼쪽에 데이터가 아무것도 없다면 
                current.left = node # 노드를 추가할 데이터 왼쪽에 넣고
                memory.append(node) # 적재
                break
            current = current.left #  왼쪽에 데이터가 있다면 왼쪽가지로 
            
        else: 
            if current.right == None: # 그 노드에 있는 숫자들 보다 크면서 오른쪽 가지가 None이라면
                current.right = node 
                memory.append(node)
                break
            current = current.right
            
print('이진 탐색 트리 구성 완료~!')

print('오늘 판매된 종류(중복x) --> ', end=' ')
preorder(root)
```
